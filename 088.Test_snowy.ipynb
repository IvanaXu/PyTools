```python
import snowy
import cv2 as cv
from IPython.display import Image
```

```python
H = 50
```

```python
Image("088.Test_snowy_view.png", height=300, width=300)
```

```python
%time img = snowy.load("088.Test_snowy_view.png")
print(img.shape)
%time img = snowy.resize(img, height=H)
print(img.shape)
%time snowy.export(img, "088.Test_snowy_out0.png")
```

```python
Image("088.Test_snowy_out0.png", height=300, width=300)
```

```python
%time img = cv.imread("088.Test_snowy_view.png")
x, y, _ = img.shape
print(img.shape)
%time img = cv.resize(img, (int(y/x*H), H))
print(img.shape)
%time cv.imwrite("088.Test_snowy_out1.png", img)
```

```python
Image("088.Test_snowy_out1.png", height=300, width=300)
```
